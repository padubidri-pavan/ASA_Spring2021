# A simple Monte Carlo code to compute the sample average of N i.i.d. exponential random variables with mean 1.

```
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

*  For sample size 1 to `samples`, generate exponential random variable of mean 1.
*  For each sample size
    *  Compute mean
    *  Compute probability, 
$\mathbb{P}[|\overline{f_N}-\pi(f)|>\delta] $, where $\overline{f_N}$ is the avearge of $N$ samples and $\pi(f)$ is the true mean of the sample.


```
samples = 10000
mean = []
probability = []
delta = 0.1
# normal = []

for i in range(1, samples+1):
    x = np.random.exponential(1, size = i)
    mean.append(x.mean())
    probability.append(min(1,(x.std())**2/(i*delta**2)))
    # normal.append((mean[i-1]-1)/i**0.5)
```

```
plt.figure(figsize=(24,6))

plt.subplot(121)
plt.title('Fig 1: Exponential Random Variable Mean vs Samples', fontsize=18)
plt.xlabel('N', fontsize=18)
plt.ylabel('$\\lambda$', fontsize=18)
plt.plot(mean, linewidth=2);
plt.hlines(y=1, xmax=samples+1, xmin=0, linewidth=2);

plt.subplot(122)
plt.title('Fig 2: $ \\mathbb{P}[|\\overline{f_N}-\\pi(f)| > \\delta] $ vs Samples', fontsize=18)
plt.xlabel('N', fontsize=18)
plt.xlabel('N', fontsize=18)
plt.plot(probability, linewidth=2);
plt.hlines(y=0, xmax=samples+1, xmin=0, linewidth=2);
```

From the `Fig 1`, it can be observed that the mean of exponential random variable is getting very close to 1. Hence the mean is 1.

From the `Fig 2`, it can be seen that as $N \to \infty$, $\mathbb{P}[|\overline{f_N}-\pi(f)| > \delta] = 0$ for $\delta > 0$. Therefore Weak law of large number is verified.
